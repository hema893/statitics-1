```python
01. What is hypothesis testing in statistics?
```

```python
**Hypothesis testing** in statistics is a method used to make decisions or inferences about a population based on sample data. It involves:

* Formulating two hypotheses:

  * **Null hypothesis (H₀):** Assumes no effect or no difference.
  * **Alternative hypothesis (H₁):** Assumes there is an effect or difference.

* Collecting sample data and using a statistical test to determine if there’s enough evidence to reject H₀ in favor of H₁.

It helps assess whether observed results are due to chance or reflect a true effect.
```

```python
02. What is the null hypothesis, and how does it differ from the alternative hypothesis?
```

```python
**Null hypothesis (H₀)** is a statement that assumes **no effect**, **no difference**, or **status quo** in the population. It is the default assumption that any observed changes are due to random chance.

**Alternative hypothesis (H₁ or Ha)** is the opposite — it suggests that **there is an effect** or **a difference**.

### Key Differences:

| Feature | Null Hypothesis (H₀)          | Alternative Hypothesis (H₁)    |
| ------- | ----------------------------- | ------------------------------ |
| Assumes | No change or difference       | A real change or difference    |
| Goal    | To test if it can be rejected | To find support for this claim |
| Example | "Drug has no effect"          | "Drug has a positive effect"   |

In hypothesis testing, we try to **reject H₀** to support H₁ based on sample data.
```

```python
03.What is the significance level in hypothesis testing, and why is it important?
```

```python
**Significance level** (denoted as **α**) in hypothesis testing is the **threshold probability** used to decide whether to reject the **null hypothesis (H₀)**.

### Key Points:

* **Common values:** 0.05 (5%), 0.01 (1%)
* It represents the **maximum risk of making a Type I error** (rejecting H₀ when it is actually true).
* If the **p-value ≤ α**, we **reject H₀**.

### Why it’s important:

It helps control how much evidence is required to make a conclusion, ensuring the results are not due to random chance. A smaller α means **stricter evidence** is needed to reject H₀.
```

```python
04.What does a P-value represent in hypothesis testing?
```

```python
A **p-value** in hypothesis testing represents the **probability of obtaining results as extreme as, or more extreme than, the observed results**, assuming that the **null hypothesis (H₀)** is true.

### Key Points:

* A **small p-value** (typically ≤ 0.05) suggests **strong evidence against H₀**, so we **reject H₀**.
* A **large p-value** (> 0.05) suggests **weak evidence against H₀**, so we **fail to reject H₀**.

### Example:

If **p = 0.03**, there is a **3% chance** the observed result is due to random chance under H₀.

### In short:

The p-value helps decide whether the evidence is strong enough to reject the null hypothesis.
```

```python
05. How do you interpret the P-value in hypothesis testing?
```

```python
A **p-value** shows the probability of getting the observed result if the **null hypothesis (H₀)** is true.

* If **p ≤ 0.05**, reject H₀ (strong evidence against H₀).
* If **p > 0.05**, do not reject H₀ (weak evidence).

It helps decide if results are due to chance.
```

```python
06.What are Type 1 and Type 2 errors in hypothesis testing?
```

```python
* **Type 1 Error (α):** Rejecting the null hypothesis when it is actually true (false positive).
* **Type 2 Error (β):** Failing to reject the null hypothesis when it is actually false (false negative).

They represent incorrect decisions in hypothesis testing.
```

```python
07. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
```

```python
* **One-tailed test:** Tests for an effect in **one specific direction** (e.g., greater than or less than).
* **Two-tailed test:** Tests for an effect in **both directions** (e.g., not equal to).

Use a one-tailed test when direction matters; two-tailed when you're checking for **any difference**.
```

```python
08. What is the Z-test, and when is it used in hypothesis testing?
```

```python
A **Z-test** is a statistical test used to determine if there is a significant difference between sample and population means when the **population standard deviation is known**.

### Used when:

* Sample size is **large** (n ≥ 30)
* Data is **normally distributed**
* **Population variance** is known

It helps test hypotheses about means or proportions.
```

```python
09. How do you calculate the Z-score, and what does it represent in hypothesis testing?
```

```python
**Z-score** is calculated using the formula:

$$
Z = \frac{{\bar{x} - \mu}}{{\sigma / \sqrt{n}}}
$$

Where:

* $\bar{x}$ = sample mean
* $\mu$ = population mean
* $\sigma$ = population standard deviation
* $n$ = sample size

### It represents:

How many **standard deviations** the sample mean is from the population mean. A higher |Z| indicates a more extreme result.
```

```python
20.What is an ANOVA test, and what are its assumptions?
```

```python
An **ANOVA (Analysis of Variance)** test compares the **means of three or more groups** to see if at least one is significantly different.

**Assumptions of ANOVA:**

1. **Independence** of observations
2. **Normality** of data within each group
3. **Equal variances** (homogeneity of variance) across groups
```

```python
21. What are the different types of ANOVA tests?
```

```python
The main types of **ANOVA tests** are:

1. **One-Way ANOVA** – Compares means of **three or more groups** based on **one independent variable**.

2. **Two-Way ANOVA** – Analyzes the effect of **two independent variables** and their **interaction** on a dependent variable.

3. **Repeated Measures ANOVA** – Used when the **same subjects** are measured multiple times under different conditions.

4. **MANOVA (Multivariate ANOVA)** – Extends ANOVA to test **multiple dependent variables** at once.
```

```python
03. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?
```

```python
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.
    
    Parameters:
    x (int): 0 or 1 (outcome)
    p (float): probability of success (1), where 0 ≤ p ≤ 1
    
    Returns:
    float: probability of observing x
    """
    if x not in [0, 1]:
        return 0
    return p if x == 1 else 1 - p

# Example usage:
print("P(X=1):", bernoulli_pdf(1, 0.7))
print("P(X=0):", bernoulli_pdf(0, 0.7))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    # Calculate mean and standard deviation
    mean = np.mean(data)
    std = np.std(data)
    
    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Plot Z-scores
    plt.scatter(range(len(data)), z_scores, color='blue')
    plt.axhline(0, color='red', linestyle='--')
    plt.title("Z-scores of Dataset")
    plt.xlabel("Data Index")
    plt.ylabel("Z-score")
    plt.grid(True)
    plt.show()
    
    return z_scores

# Example usage
data = np.random.normal(50, 10, 100)  # simulate normal data
z = calculate_and_plot_z_scores(data)
```

```python
15. Implement hypothesis testing using Z-statistics for a sample dataset?
```

```python
from scipy.stats import norm
import numpy as np

# Given sample data
sample_mean = 52.5
population_mean = 50
sample_std = 5
n = 30

# Step 1: Calculate Z-statistic
z = (sample_mean - population_mean) / (sample_std / np.sqrt(n))

# Step 2: Calculate p-value for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z)))

# Step 3: Set significance level
alpha = 0.05

# Step 4: Output results
print(f"Z-statistic: {z:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("✅ Reject the null hypothesis (significant difference).")
else:
    print("❌ Fail to reject the null hypothesis (no significant difference).")
```

```python
16. Create a confidence interval for a dataset using Python and interpret the result.
```

```python
import numpy as np
import scipy.stats as stats

# Sample dataset
data = [52, 55, 48, 50, 53, 54, 49, 51, 52, 56]

# Step 1: Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)

# Step 2: Set confidence level
confidence = 0.95
z_critical = stats.norm.ppf(1 - (1 - confidence) / 2)  # for 95%, z ≈ 1.96

# Step 3: Compute margin of error and confidence interval
margin_of_error = z_critical * (sample_std / np.sqrt(n))
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Step 4: Display results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
24.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distrubation.
```

```python
import numpy as np
from scipy.stats import skew, kurtosis

def analyze_distribution(size=1000, mean=0, std=1):
    # Generate data from a normal distribution
    data = np.random.normal(loc=mean, scale=std, size=size)
    
    # Calculate skewness and kurtosis
    data_skewness = skew(data)
    data_kurtosis = kurtosis(data)  # default is excess kurtosis
    
    # Display results
    print(f"Skewness: {data_skewness:.4f}")
    print(f"Excess Kurtosis: {data_kurtosis:.4f}")
    
    return data_skewness, data_kurtosis

# Call the function
analyze_distribution()
```

```python
01. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret and result.
```

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample = [52, 49, 51, 53, 54, 48, 50, 55, 47, 50]
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)  # sample standard deviation
n = len(sample)

# Known population mean
population_mean = 50

# Step 1: Calculate Z-statistic
z_stat = (sample_mean - population_mean) / (sample_std / np.sqrt(n))

# Step 2: Calculate p-value for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Step 3: Output results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Step 4: Interpret at significance level alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("✅ Reject the null hypothesis: Sample mean is significantly different from the population mean.")
else:
    print("❌ Fail to reject the null hypothesis: No significant difference from the population mean.")
```

```python
04. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
```

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Step 1: Simulated sample data
sample = [102, 98, 100, 101, 99, 97, 103, 96, 104, 98]
n = len(sample)
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)

# Step 2: Known population mean
population_mean = 100

# Step 3: Compute Z-statistic
z_stat = (sample_mean - population_mean) / (sample_std / np.sqrt(n))

# Step 4: Calculate p-value (two-tailed)
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Step 5: Critical Z-values for α = 0.05
alpha = 0.05
z_critical = norm.ppf(1 - alpha / 2)  # ≈ ±1.96 for 95% CI

# Step 6: Print test result
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_stat:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Critical Z-values: ±{z_critical:.2f}")

if abs(z_stat) > z_critical:
    print("✅ Reject the null hypothesis: Significant difference.")
else:
    print("❌ Fail to reject the null hypothesis: No significant difference.")

# Step 7: Visualization
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Standard Normal Distribution', color='blue')

# Shade rejection regions
plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.5, label='Rejection Region (Left)')
plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.5, label='Rejection Region (Right)')

# Plot z-statistic line and critical lines
plt.axvline(z_stat, color='black', linestyle='--', label=f'Z = {z_stat:.2f}')
plt.axvline(z_critical, color='gray', linestyle=':', label=f'+Z_critical = {z_critical:.2f}')
plt.axvline(-z_critical, color='gray', linestyle=':', label=f'-Z_critical = {-z_critical:.2f}')

# Plot formatting
plt.title('Two-Tailed Z-Test with Rejection Regions')
plt.xlabel('Z-value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
06. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2(mu0, mu1, sigma, n, alpha=0.05):
    """
    Visualizes Type I and Type II errors for a Z-test.
    
    Parameters:
    - mu0: Null hypothesis mean
    - mu1: True mean (alternative hypothesis)
    - sigma: Population standard deviation
    - n: Sample size
    - alpha: Significance level (default = 0.05)
    """

    # Calculate standard error
    se = sigma / np.sqrt(n)

    # Calculate critical value for right-tailed test
    z_critical = norm.ppf(1 - alpha)
    x_critical = mu0 + z_critical * se

    # Calculate beta (Type II error)
    z_beta = (x_critical - mu1) / se
    beta = norm.cdf(z_beta)

    # Create x range for plotting
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)

    # PDF under H0 and H1
    y0 = norm.pdf(x, mu0, se)
    y1 = norm.pdf(x, mu1, se)

    # Plot distributions
    plt.figure(figsize=(10, 5))
    plt.plot(x, y0, label='H₀: μ = {}'.format(mu0), color='blue')
    plt.plot(x, y1, label='H₁: μ = {}'.format(mu1), color='green')

    # Shade Type I error region (right tail of H0)
    plt.fill_between(x, y0, where=(x >= x_critical), color='red', alpha=0.4, label='Type I Error (α)')

    # Shade Type II error region (left of x_critical under H1)
    plt.fill_between(x, y1, where=(x <= x_critical), color='orange', alpha=0.4, label='Type II Error (β)')

    # Critical value line
    plt.axvline(x_critical, color='black', linestyle='--', label=f'Critical value = {x_critical:.2f}')

    # Labels and legend
    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Output numerical results
    print(f"Critical Value (x_critical): {x_critical:.2f}")
    print(f"Type I Error (α): {alpha:.2f}")
    print(f"Type II Error (β): {beta:.4f}")

# Example usage
visualize_type1_type2(mu0=100, mu1=105, sigma=15, n=30, alpha=0.05)
```

```python
06. Write a Python program to perform an independent T-test and interpret the results.
```

```python
import numpy as np
from scipy.stats import ttest_ind

# Step 1: Sample data (e.g., test scores from two groups)
group1 = [88, 92, 85, 91, 87, 90, 86, 89]
group2 = [84, 83, 85, 87, 82, 80, 86, 84]

# Step 2: Perform independent two-sample T-test
t_stat, p_value = ttest_ind(group1, group2, equal_var=False)  # Welch's t-test

# Step 3: Output results
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Step 4: Interpret results at significance level α = 0.05
alpha = 0.05
if p_value < alpha:
    print("✅ Reject the null hypothesis: Significant difference between the groups.")
else:
    print("❌ Fail to reject the null hypothesis: No significant difference between the groups.")
```

```python
import numpy as np
from scipy.stats import ttest_rel
import matplotlib.pyplot as plt

# Step 1: Paired sample data (e.g., test scores before and after training)
before = [70, 68, 75, 71, 69, 74, 72, 70]
after  = [74, 70, 78, 73, 72, 76, 75, 73]

# Step 2: Perform paired t-test
t_stat, p_value = ttest_rel(before, after)

# Step 3: Print test results
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Step 4: Interpret results
alpha = 0.05
if p_value < alpha:
    print("✅ Reject the null hypothesis: Significant difference between paired samples.")
else:
    print("❌ Fail to reject the null hypothesis: No significant difference between paired samples.")

# Step 5: Visualization
x = np.arange(len(before))
width = 0.35

plt.figure(figsize=(8, 5))
plt.bar(x - width/2, before, width, label='Before', color='skyblue')
plt.bar(x + width/2, after, width, label='After', color='salmon')
plt.xticks(x, [f'Pair {i+1}' for i in x])
plt.ylabel('Scores')
plt.title('Paired Sample Comparison (Before vs After)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from scipy.stats import ttest_1samp, norm

# Step 1: Simulate data from a normal distribution
np.random.seed(42)
population_mean = 50
population_std = 10  # known std for Z-test
sample = np.random.normal(loc=52, scale=10, size=15)  # small sample size

# Step 2: Z-test (assuming known population std)
sample_mean = np.mean(sample)
n = len(sample)
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))
z_p_value = 2 * (1 - norm.cdf(abs(z_stat)))  # two-tailed

# Step 3: T-test (sample std used)
t_stat, t_p_value = ttest_1samp(sample, population_mean)

# Step 4: Display results
print("=== Z-Test ===")
print(f"Z-statistic: {z_stat:.2f}")
print(f"P-value: {z_p_value:.4f}")

print("\n=== T-Test ===")
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {t_p_value:.4f}")

# Step 5: Interpretation
alpha = 0.05
print("\n=== Conclusion ===")
if z_p_value < alpha:
    print("✅ Z-test: Reject the null hypothesis.")
else:
    print("❌ Z-test: Fail to reject the null hypothesis.")

if t_p_value < alpha:
    print("✅ T-test: Reject the null hypothesis.")
else:
    print("❌ T-test: Fail to reject the null hypothesis.")
```

```python
09.Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
```

```python
import scipy.stats as stats
import numpy as np

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the mean of a sample.

    Parameters:
    - data (list or array): The sample data.
    - confidence (float): The confidence level (default is 0.95 for 95%).

    Returns:
    - tuple: (mean, lower bound, upper bound)
    """
    sample = np.array(data)
    n = len(sample)
    mean = np.mean(sample)
    std_err = stats.sem(sample)  # Standard error of the mean
    margin = std_err * stats.t.ppf((1 + confidence) / 2, df=n-1)  # t-distribution for small samples

    return mean, mean - margin, mean + margin
```

```python
10. Write a Python program to calculate the margin of error for a given confidence level using sample data.
```

```python
import scipy.stats as stats
import numpy as np

def margin_of_error(data, confidence=0.95):
    n = len(data)
    std_err = stats.sem(data)  # Standard error
    critical_value = stats.t.ppf((1 + confidence) / 2, df=n-1)
    return critical_value * std_err

# Example usage
sample = [45, 50, 47, 52, 49, 48, 46]
moe = margin_of_error(sample, 0.95)
print(f"Margin of Error: {moe:.2f}")
```

```python
11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
```

```python
def bayes_theorem(prior_H, likelihood_D_given_H, likelihood_D_given_not_H):
    """
    Apply Bayes' Theorem to compute P(H|D)
    """
    prior_not_H = 1 - prior_H
    # Compute marginal likelihood P(D)
    evidence = (likelihood_D_given_H * prior_H) + (likelihood_D_given_not_H * prior_not_H)
    
    # Compute posterior P(H|D)
    posterior = (likelihood_D_given_H * prior_H) / evidence
    return posterior

# Given:
P_H = 0.01                  # Prior: Probability of disease
P_D_given_H = 0.99          # Likelihood: Test is positive given disease
P_D_given_not_H = 1 - 0.95  # False positive rate: Test is positive without disease

# Compute posterior probability
posterior_prob = bayes_theorem(P_H, P_D_given_H, P_D_given_not_H)
print(f"Probability of having disease given a positive test: {posterior_prob:.4f}")
```

```python
12. Perform a Chi-square test for independence between two categorical variables in Python.
```

```python
import numpy as np
from scipy.stats import chi2_contingency

# Create contingency table
data = np.array([[30, 10],  # Tea
                 [20, 40]]) # Coffee

# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(data)

# Display results
print(f"Chi-square statistic = {chi2:.4f}")
print(f"Degrees of freedom = {dof}")
print(f"P-value = {p:.4f}")
print("Expected frequencies:\n", expected)

# Interpretation
if p < 0.05:
    print("Result: Reject null hypothesis - variables are dependent.")
else:
    print("Result: Fail to reject null hypothesis - variables are independent.")
```

```python
13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.
```

```python
import numpy as np

# Sample observed data (contingency table)
# Rows: categories of variable A
# Columns: categories of variable B
observed = np.array([
    [30, 10],  # Category A1
    [20, 40]   # Category A2
])

# Calculate row totals and column totals
row_totals = observed.sum(axis=1).reshape(-1, 1)  # Convert to column vector
col_totals = observed.sum(axis=0).reshape(1, -1)  # Convert to row vector
grand_total = observed.sum()

# Calculate expected frequencies
expected = (row_totals @ col_totals) / grand_total  # Matrix multiplication

# Display results
print("Observed Frequencies:\n", observed)
print("Expected Frequencies:\n", expected.round(2))
```

```python
14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
```

```python
from scipy.stats import chisquare

# Observed frequencies from the experiment
observed = [8, 9, 10, 11, 12, 10]

# Expected frequencies for a fair die (uniform distribution)
expected = [10, 10, 10, 10, 10, 10]

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Output results
print(f"Chi-square statistic = {chi2_stat:.4f}")
print(f"P-value = {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("Result: Reject null hypothesis - distribution differs significantly from expected.")
else:
    print("Result: Fail to reject null hypothesis - no significant difference from expected.")
```
